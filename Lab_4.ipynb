```python
#import necessary tools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv("files_for_lab/marketing_customer_analysis.csv")
df.head()
```

```python
df_num = df.select_dtypes(include=[np.number])
df_num.head()
```

```python
#checkcorrelation
pearson = sns.heatmap(df_num.corr(), annot = True)
```

```python
kendall = sns.heatmap(df_num.corr(method = "kendall"), annot = True)
```

```python
spearman = sns.heatmap(df_num.corr(method = "spearman"), annot = True)
```

```python
#check for normality
from scipy.stats import skew
df_num.apply(skew)
from scipy.stats import kurtosis
kurtosis(df_num)
from statsmodels.graphics.gofplots import qqplot
cols = list(df_num.columns)
for col in cols:
    print("Q-Q plot for variable: ",col)
    qqplot(df_num[col], line='s')
    plt.show()
```
